```python
import sqlite3
import pandas as pd
pd.set_option('display.max_colwidth',None)
db_path = r"C:\Users\pandu\OneDrive\Pictures\Documents\trend_pulse\youtube_data.db"
conn = sqlite3.connect(db_path)
df = pd.read_sql_query("SELECT * FROM trending_videos",conn)
def analyze_data(df):
    """Perform analysis and return insights"""

    insights = {}
    df["title_short"] = df["title"].str.slice(0,30)
    df["like_ratio"] = df["likes"] / df["views"]
    # Top videos
    insights["top_views"] = df.nlargest(10, "views")[["title_short", "views"]]
    insights["top_likes"] = df.nlargest(10, "likes")[["title_short", "likes"]]
    insights["top_engagement"] = df.nlargest(10, "like_ratio")[["title_short", "like_ratio"]]

    # Summary stats
    insights["summary"] = df.describe()

    # Best video
    insights["best_video"] = df.loc[df["views"].idxmax()]

    return insights

# Run analysis
results = analyze_data(df)

print("🔥 Top Videos by Views:")
print(results["top_views"])

print("\n📊 Summary:")
print(results["summary"])
```
